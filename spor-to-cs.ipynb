```python
import json
from os.path import exists
from os import listdir, getcwd, mkdir, system, rmdir, removedirs, remove
import subprocess
import shutil
import pandas as pd
```

```python
DICT_FOLDER = "SPOR-Lists-EE/"

dict_data = {}
for file in listdir(DICT_FOLDER):
   if file=="eHDSI-Substances-EEmappings.xlsx":
        continue
 #   print(file)
   n_file = file.split(".")[0]
    # df = pd.read_excel(
    #     "Amlodipine/SPOR-Lists-EE/" + str(data_dictionary) + ".xlsx", header=2
    # )
   df = pd.read_excel(DICT_FOLDER + "/" + file, header=2)
   dict_data[n_file] = df
```

```python
for element in dict_data.keys():

   # print(element)
    cs = {
    "resourceType": "CodeSystem",
    "meta": {
        "lastUpdated": "2020-04-09T21:10:28.568+00:00",
        "profile": [
            "http://hl7.org/fhir/StructureDefinition/shareablecodesystem"
        ]
    },
    "identifier": [
        {
            "system": "urn:ietf:rfc:3986",
            "value": "urn:oid:2.16.840.1.113883.4.642.1.1147"
        }
    ],
    "version": "1.0.0",
    
    "status": "draft",
    "experimental": False,
    "date": "2023-11-17T05:18:48-07:00",
    "publisher": "Health Level Seven International",
    "contact": [
        {
            "telecom": [
                {
                    "system": "url",
                    "value": "http://hl7.org"
                },
                {
                    "system": "email",
                    "value": "hq@HL7.org"
                }
            ]
        }
    ],
    "description": "This value set includes Coverage Class codes.",
    "copyright": "This material derives from the HL7 Terminology (THO). THO is copyright Â©1989+ Health Level Seven International and is made available under the CC0 designation. For more licensing information see: https://terminology.hl7.org/license",
    "caseSensitive": True,
    "content": "complete",
    "concept": []
}
    cs["id"]=element
    cs["url"]= "https://spor.ema.europa.eu/v1/lists/"+element
    cs["name"]= "SPOR "+element
    cs["title"]= "SPOR "+element
    cs["valueSet"]= "http://terminology.hl7.org/ValueSet/"+element


    for row in dict_data[element].iterrows():
        tt={}
       # print(row[1]["RMS termini id"])
       # print(row[1]["RMS termini nimi"])
       # print(row[1]["Kirjeldus"])
        
        tt["code"]=str(row[1]["RMS termini id"])
        tt["display"]=row[1]["RMS termini nimi"]
        if not pd.isna(row[1]["Kirjeldus"]):
                tt["definition"]=row[1]["Kirjeldus"]
               
        cs["concept"].append(tt)

    with open(element+".json", "w") as outfile:  # save as a json just because
        outfile.write(json.dumps(cs, indent=2))
```
